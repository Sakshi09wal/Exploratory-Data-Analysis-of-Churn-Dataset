The first categorical variable we investigate is International Plan. A bar

```python
df["Int'l Plan"].value_counts()
```

 When observing categorical variables, it is often a useful practice to create contingency tables for distributions and marginal
distributions, in this case, to quantify the relationship between Churn (T/F) and International Plan (Y/N). These can be graphed (and overlayed) to show proportions.

```python
inttable = pd.crosstab(df["Int'l Plan"], df["Churn"])
display(inttable )
```

```python
df.groupby(["Int'l Plan", "Churn"]).size().unstack().plot(kind='bar', stacked=True, figsize=(5,5),title = "Churn Proportions by International Plan")
plt.xlabel('International Plan')
plt.ylabel('Count')
plt.show()
```

Notably, A bar chart of the International Plan, with an overlay of churn, and represents a comparison of the proportion of churners and nonchurners,among customers who either had selected the International Plan (yes, 323 customers) or had not selected it (no, 3010 customers).

```python
#Table of Sums for Both Variables
intsum = pd.crosstab(df["Int'l Plan"], df["Churn"], margins = True)
display(intsum)
```

```python
#Table of proportions
intprop = intsum / intsum.ix["All", "All"]
display(intprop)
```

# 7) Stacked Bar of Customer Service Calls and Churn

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
cust_churn = pd.crosstab(df['CustServ Calls'], df['Churn'])
cust_churn.plot(kind = 'barh', stacked = True)
plt.ylabel('Count')
plt.xlabel('Calls to Customer Service')
plt.title('Churn Rate Relative to Calls to Customer Service')
plt.show()
```

This graph is interesting. Once a customer gets up to 4 customer service calls, the churn rate drastically increases. It is a little bit difficult to see churn for 7 or more customer service calls. Therefore, we can normalize the bar graph so that theyâ€™re all equal height, but the proportions all remain the same. That way we can get a real grasp for what the churn rate is for the customers that make 7 or more customer service calls.

# 7) Normalized histogram of customer service calls with churn overlay.

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
cust_norm = cust_churn.div(cust_churn.sum(1).astype(float), axis = 0)
cust_norm.plot(kind = 'bar', stacked = True, title = "Normalized Churn Rate by Customer Service Calls")
plt.xlabel("Customer Service Calls")
plt.ylabel("Percent")
plt.show()
```

We therefore turn to the â€œnormalizedâ€ histogram, where every rectangle has the same height and width.The pattern now becomes crystal clear. Customers who have called customer
service three times or less have a markedly lower churn rate (red part of the rectangle) than customers who have called customer service four or more times.

# 9) Scatterplot of Evening Minutes and Day Minutes by Churn

```python
seaborn.lmplot('Eve Mins', 'Day Mins', data = df, hue = 'Churn', fit_reg = False, legend_out = False)
plt.title('Scatterplot of Day and Evening Minutes by Churn')
plt.xlabel("Evening Minutes")
plt.ylabel("Day Minutes")
plt.show()
```

records below the line.

Two variables x and y are linearly correlated if an increase in x is associated with either an increase in y or a decrease in y. The
correlation coefficient r quantifies the strength and direction of the linear relationship between x and y. The threshold for significance of
the correlation coefficient r depends not only on the sample size but also on data mining, where there are a large number of records (over
1000), even small values of r, such as may be statistically significant.At best, using correlated variables will overemphasize one data
component; at worst, using correlated variables will cause the model to become unstable and deliver unreliable results. However, just
because two variables are correlated does not mean that we shouldomit one of them.
